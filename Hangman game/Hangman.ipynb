```python
top = Tk()

#Customising window
w = 200 #width for the root
h = 200 #height for the root
ws = top.winfo_screenwidth()#width for the screen
hs = top.winfo_screenheight()#height for the screen
# coordinates for the window
x = (ws/2)-(w/2)
y = (hs/2)-(h/2)
# setting the position and dimesions of the window 
top.geometry('%dx%d+%d+%d' % (w,h,x,y))

#Get Player1's name
Player1_label = Label(top, text="Player 1 is")
Player1_label.pack()
Player1_textbox = Entry(top, bd=1)
Player1_textbox.pack()

#Get Player2's name
Player2_label = Label(top, text='Player 2 is')
Player2_label.pack()
Player2_textbox = Entry(top, bd=1)
Player2_textbox.pack()

#Enter button
def store_input():
    global Player1
    global Player2
    Player1 = Player1_textbox.get()
    Player2 = Player2_textbox.get()


enter_button = Button(top, text="Enter", command = store_input)
enter_button.pack(side="left")

#Close button

close_button = Button(top, text = "Close", command = top.destroy)
close_button.pack(side = 'right')

top.mainloop()
```
